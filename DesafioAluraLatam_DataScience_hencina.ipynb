```python
import pandas as pd

#venta total por tienda

# Cargar datos
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Leer CSVs
tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Calcular ingreso total por tienda (sumando columna Precio)
ingreso_tienda_1 = tienda['Precio'].sum()
ingreso_tienda_2 = tienda2['Precio'].sum()
ingreso_tienda_3 = tienda3['Precio'].sum()
ingreso_tienda_4 = tienda4['Precio'].sum()

# Crear resumen
ingresos_totales = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ingreso Total': [ingreso_tienda_1, ingreso_tienda_2, ingreso_tienda_3, ingreso_tienda_4]
})

# Mostrar resultados
print(ingresos_totales)
```

```python
import matplotlib.pyplot as plt

# Paleta de colores para cada punto de la línea
colores = ['#94d2bd', '#005f73', '#0a9396', '#ee9b00']

# Gráfico
plt.figure(figsize=(8, 6))

# Gráfico de líneas
plt.plot(
    ingresos_totales['Tienda'],
    ingresos_totales['Ingreso Total'],
    marker='o',
    linestyle='-',
    linewidth=2.5,
    color='#005f73'
)

# Colores por cada punto
for i, color in enumerate(colores):
    plt.plot(
        ingresos_totales['Tienda'][i],
        ingresos_totales['Ingreso Total'][i],
        marker='o',
        markersize=10,
        color=color,
        label=ingresos_totales['Tienda'][i]
    )

# Título y etiquetas
plt.title('Ingreso Total por Tienda', fontsize=16, fontweight='bold')
plt.ylabel('Ingreso ($)', fontsize=12)
plt.xlabel('Tienda', fontsize=12)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# leyenda
plt.legend(title='Tienda')

# Mostrar gráfico
plt.tight_layout()
plt.show()

```

```python
import pandas as pd


url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Cargar los datos
tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# limpieza de los datos
for df in [tienda1, tienda2, tienda3, tienda4]:
    df.columns = df.columns.str.strip()

# se crear función para contar productos por categoría
def contar_por_categoria(df, nombre_tienda):
    conteo = df['Categoría del Producto'].value_counts().reset_index()
    conteo.columns = ['Categoria', 'Cantidad']
    conteo['Tienda'] = nombre_tienda
    return conteo

# acá aplico la función a cada tienda
cat_tienda1 = contar_por_categoria(tienda1, 'Tienda 1')
cat_tienda2 = contar_por_categoria(tienda2, 'Tienda 2')
cat_tienda3 = contar_por_categoria(tienda3, 'Tienda 3')
cat_tienda4 = contar_por_categoria(tienda4, 'Tienda 4')

# en este código uno los resultados en un solo DataFrame
ventas_categorias = pd.concat([cat_tienda1, cat_tienda2, cat_tienda3, cat_tienda4])

# lueego mostrar resultados
print(ventas_categorias)


```

# **5. Productos mas vendidos por cada tienda**
Busco saber cuál tienda genera más ventas a partir de sus productos más vendidos, es decir:

Tomar los productos más vendidos por tienda (basado en cantidad de ventas).

Calcular el valor total en dinero que generan esos productos en cada tienda.

Comparar entre tiendas cuál genera más ingresos con sus productos top.


```python
import pandas as pd

# Cargar datos
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Etiquetar cada tienda
tienda1 = pd.read_csv(url1)
tienda1['Tienda'] = 'Tienda 1'

tienda2 = pd.read_csv(url2)
tienda2['Tienda'] = 'Tienda 2'

tienda3 = pd.read_csv(url3)
tienda3['Tienda'] = 'Tienda 3'

tienda4 = pd.read_csv(url4)
tienda4['Tienda'] = 'Tienda 4'

# Unir todas las tiendas
df = pd.concat([tienda1, tienda2, tienda3, tienda4])

# Calcular los productos más vendidos por tienda (por cantidad de ocurrencias)
ventas_por_producto = df.groupby(['Tienda', 'Producto']).agg(
    cantidad_ventas=('Producto', 'count'),
    total_ventas=('Precio', 'sum')
).reset_index()

# Obtener los productos más vendidos por cada tienda (top 5 por ejemplo)
top_productos_por_tienda = ventas_por_producto.sort_values(
    ['Tienda', 'cantidad_ventas'], ascending=[True, False]
).groupby('Tienda').head(5)

# Calcular el total que generan esos productos top por tienda
ventas_por_tienda_top_productos = top_productos_por_tienda.groupby('Tienda')['total_ventas'].sum().reset_index()

# Formatear con separador de miles y 2 decimales
ventas_por_tienda_top_productos['total_ventas'] = ventas_por_tienda_top_productos['total_ventas'].apply(lambda x: f"${x:,.2f}")

# Mostrar resultado
print("✅ Ventas totales generadas por los productos más vendidos de cada tienda:")
print(ventas_por_tienda_top_productos)
```

```python
import pandas as pd

# Cargar los datos desde las URLs
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Leer y etiquetar cada tienda
tienda1 = pd.read_csv(url1)
tienda1['Tienda'] = 'Tienda 1'

tienda2 = pd.read_csv(url2)
tienda2['Tienda'] = 'Tienda 2'

tienda3 = pd.read_csv(url3)
tienda3['Tienda'] = 'Tienda 3'

tienda4 = pd.read_csv(url4)
tienda4['Tienda'] = 'Tienda 4'

# Unir todos los datos en un solo DataFrame
df_tiendas = pd.concat([tienda1, tienda2, tienda3, tienda4])

# Agrupar por tienda y calcular totales
resumen = df_tiendas.groupby('Tienda').agg({
    'Precio': 'sum',
    'Costo de envío': 'sum'
}).reset_index()

# Renombrar columnas
resumen.columns = ['Tienda', 'Valor Total Ventas', 'Valor Total Costo Envío']

# Calcular ganancia aproximada
resumen['Ganancia (Aprox)'] = resumen['Valor Total Ventas'] - resumen['Valor Total Costo Envío']

# Formatear con separadores de miles y 2 decimales
resumen_formateado = resumen.copy()
resumen_formateado[['Valor Total Ventas', 'Valor Total Costo Envío', 'Ganancia (Aprox)']] = resumen_formateado[
    ['Valor Total Ventas', 'Valor Total Costo Envío', 'Ganancia (Aprox)']
].applymap(lambda x: f"${x:,.2f}")

# Mostrar tabla final formateada
print(resumen_formateado)

```

# Análisis de Datos de Tiendas para Decisión de Venta

## Descripción del Proyecto

Este proyecto analiza los datos de ventas de cuatro tiendas diferentes para determinar cuál de ellas es la candidata más adecuada para ser vendida. El análisis se basa en diversas métricas de rendimiento, como ingresos, ventas por categoría, calificaciones de clientes y costos de envío.

## Archivos de Datos

El análisis utiliza cuatro archivos CSV, cada uno correspondiente a una tienda:

*   `tienda_1.csv`
*   `tienda_2.csv`
*   `tienda_3.csv`
*   `tienda_4.csv`

Cada archivo contiene información detallada sobre las transacciones, incluyendo:

*   Producto
*   Categoría del Producto
*   Precio
*   Costo de envío
*   Fecha de Compra
*   Vendedor
*   Lugar de Compra
*   Calificación
*   Método de pago
*   Cantidad de cuotas
*   Latitud y Longitud del cliente

## Análisis Realizado

Se llevaron a cabo los siguientes análisis para evaluar el rendimiento de cada tienda:

1.  **Análisis de Facturación:** Se calcularon y compararon los ingresos totales de cada tienda.
2.  **Ventas por Categoría:** Se identificaron las categorías de productos más vendidas en cada tienda para entender la demanda de los clientes.
3.  **Calificación Promedio de la Tienda:** Se analizó la satisfacción del cliente a través de las calificaciones promedio de los productos.
